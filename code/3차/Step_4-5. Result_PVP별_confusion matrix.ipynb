```python
df_unlabeled = pd.read_csv("C:/Users/user/Desktop/논문리뷰/Pattern-Exploiting-Training(PET)/code/testPET - 복사본/data3/df_unlabeled.csv")
```

```python
unlabel_array = np.array(df_unlabeled["label"])
```

```python
unlabel_array= np.where(unlabel_array == 1, 0, np.where(unlabel_array == 2, 1, unlabel_array))
```

```python
all_array = np.array([PVP0_array,
                      PVP1_array,
                      PVP2_array,
                      PVP3_array,
                      PVP4_array,
                      PVP5_array,
                      PVP6_array,
                      PVP7_array,
                      PVP8_array,
                      PVP9_array])
```

```python
unlabel_array
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
cm_array_2d = []
for i in all_array:
    cm = confusion_matrix(unlabel_array, i)
    cm_array_2d.append(cm)
```

```python
cm_array_2d
```

```python
cmap = plt.cm.get_cmap("Blues")
```

```python
for i in cm_array_2d:
    cm_display = ConfusionMatrixDisplay(i).plot(cmap=cmap)
```

```python
cm_columns = ["TN", "FP", "FN", "TP"]
```

```python
all_index = ['PVP0',
           'PVP1',
           'PVP2',
           'PVP3',
           'PVP4',
           'PVP5',
           'PVP6',
           'PVP7',
           'PVP8',
           'PVP9']
```

```python
cm_list_1d = []
for i in cm_array_2d:
    cm_1d = i.reshape(-1)
    cm_list_1d.append(cm_1d)   
```

```python
df_cm = pd.DataFrame(cm_list_1d, columns=cm_index).set_index(pd.Index(all_index))
```
